```python
for col in numeric_cols_names:
  col_outliers_index = detect_outliers(df, 0 , [col])
  col_median = df[col].median()
  df[col].iloc[col_outliers_index] = col_median
```

```python
# prompt: APPLY One hot encoder in this data  doornumber and  aspiration and fueltype and enginelocation and drivewheel
from sklearn.preprocessing import OneHotEncoder
# Create the encoder
encoder = OneHotEncoder()

# Fit and transform the specified columns
columns_to_encode = ['doornumber', 'aspiration', 'fueltype', 'enginelocation', 'drivewheel','cylindernumber','carbody']
encoded_data = encoder.fit_transform(df[columns_to_encode])

# Create a DataFrame from the encoded data
encoded_df = pd.DataFrame(encoded_data.toarray(), columns=encoder.get_feature_names_out(columns_to_encode))

# Concatenate the encoded DataFrame with the original DataFrame
df = pd.concat([df, encoded_df], axis=1)

# Drop the original columns
df = df.drop(columns_to_encode, axis=1)

# Display the updated DataFrame
df.head()
```

```python
# prompt: apply regression model for this df

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Assuming 'price' is the target variable
X = df.drop('price', axis=1)
y = df['price']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=20, random_state=42)

# Create a linear regression model
model = LinearRegression()

# Fit the model to the training data
model.fit(X_train, y_train)

# Make predictions on the test data
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print('Mean Squared Error:', mse)
print('R-squared:', r2)
```
